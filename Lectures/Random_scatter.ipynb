```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate 2D data with different variance in two directions
def generate_rotated_data(mean, cov, n_points):
    data = np.random.multivariate_normal(mean, cov, n_points)
    return data

# Mean of the data points
mean = [0, 0]

# Covariance matrix with different variances (not aligned with x or y axis)
cov = [[2.0, 0.8],   # Variance along x' and y' (rotated axes)
       [0.8, 0.6]] 

# Generate 10 points with this distribution
data = generate_rotated_data(mean, cov, 10)

```

```python
# Plot the points
fig, ax = plt.subplots(figsize=(6,6))

# Scatter first 2 points (in red, circles)
ax.scatter(data[:, 0], data[:, 1], color='red', s=100, alpha=0.3, marker='o')
ax.quiver(0, 0, -1, 1, angles='xy', scale_units='xy', scale=1, color='green', label='Vector (1, -1)', width=0.005)
# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
#ax.set_title('Scatter Plot with Different Variance and Rotation')

# Add legend
#ax.legend()

# Show plot
plt.grid(True)
plt.savefig('vec1.png')
```

```python
def multiply_vec(cov_matrix, vector):
    return np.dot(cov_matrix, vector)
```
