```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function f(x) = ax^2 + bx + c
def f(x):
    return a * x**2 + b * x + c

# Define the derivative (slope of the tangent) of the quadratic function
def f_prime(x):
    return 2 * a * x + b

# Get the equation of the tangent line at a point x0
def tangent_line(x0, x):
    return f(x0) + f_prime(x0) * (x - x0)

# Coefficients for the quadratic function
a = 1
b = -2
c = 1

# Generate x values for the plot
x = np.linspace(-10, 10, 400)
y = f(x)

# Select two random points where we will calculate the tangents
np.random.seed(5)  # For reproducibility
x_tangent_points = np.random.uniform(-5, 5, 2)

# Calculate the y values at the tangent points
y_tangent_points = f(x_tangent_points)

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the quadratic function
plt.plot(x, y, label=f'Quadratic Function: $y = {a}x^2 + {b}x + {c}$', color='purple')

# Plot the tangent lines at the two random points
for x0 in x_tangent_points:
    plt.plot(x, tangent_line(x0, x), '--', label=f'Tangent at x={x0:.2f}')

# Mark the points of tangency
plt.scatter(x_tangent_points, y_tangent_points, color="#006400", zorder=5)

# Add labels and title
#plt.title('Quadratic Function with Tangents at Two Points')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-4,100)

# Add a legend
#plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
