```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate some noisy data points
np.random.seed(15)
x = np.linspace(-7, 3, 50)
y = (x + 5) * (x + 2) * (x - 1) + np.random.normal(0, 12, 50)

# Define the polynomial functions you want to fit
def linear_polynomial(x, a, b):
    return a * x + b

def quadratic_polynomial(x, a, b, c):
    return a * x**2 + b * x + c

def cubic_polynomial(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

def quartic_polynomial(x, a, b, c, d, e):
    return a * x**4 + b * x**3 + c * x**2 + d * x + e

def quintic_polynomial(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

def sixth_order(x, a, b, c, d, e, f, g):
    return a * x**6 + b * x**5 + c * x**4 + d * x**3 + e * x**2 + f * x + g

# Fit polynomials of different orders to the noisy data
params_linear, _ = curve_fit(linear_polynomial, x, y)
params_quadratic, _ = curve_fit(quadratic_polynomial, x, y)
params_cubic, _ = curve_fit(cubic_polynomial, x, y)
params_quartic, _ = curve_fit(quartic_polynomial, x, y)
params_quintic, _ = curve_fit(quintic_polynomial, x, y)
params_sixth, _ = curve_fit(sixth_order, x, y)

# Create a smoother x-axis for the fitted curves
x_smooth = np.linspace(-8.5, 4, 100)

# Calculate the corresponding y-values for the fitted polynomial curves
y_linear_fit = linear_polynomial(x_smooth, *params_linear)
y_quadratic_fit = quadratic_polynomial(x_smooth, *params_quadratic)
y_cubic_fit = cubic_polynomial(x_smooth, *params_cubic)
y_quartic_fit = quartic_polynomial(x_smooth, *params_quartic)
y_quintic_fit = quintic_polynomial(x_smooth, *params_quintic)
y_sixth_fit = sixth_order(x_smooth, *params_sixth)

# Plot the original data and the fitted polynomial curves
plt.scatter(x, y, label='Noisy Data', color='k')
plt.plot(x_smooth, y_linear_fit, 'violet', label='1st order')
plt.plot(x_smooth, y_quadratic_fit, 'teal', label='2nd order')
plt.plot(x_smooth, y_cubic_fit, 'orange', label='3rd order')
plt.plot(x_smooth, y_quartic_fit, 'lightcoral', label='4th order')
plt.plot(x_smooth, y_quintic_fit, 'deepskyblue', label='5th order')
plt.plot(x_smooth, y_sixth_fit, 'forestgreen', label='6th order')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Fitting Polynomials of Different Orders to Noisy Data')
plt.grid(True)
plt.ylim(-120,120)
plt.xlim(-9,5)
plt.savefig('data6.jpg')
#plt.show()

# Print the fitted coefficients for each polynomial
#print("Fitted coefficients for Linear Polynomial:", params_linear)
#print("Fitted coefficients for Quadratic Polynomial:", params_quadratic)
#print("Fitted coefficients for Cubic Polynomial:", params_cubic)
#print("Fitted coefficients for Quartic Polynomial:", params_quartic)
#print("Fitted coefficients for Quintic Polynomial:", params_quintic)
```
